# Multiple Linear Regression

## Importing the Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

## Importing the Dataset

```python
dataset = pd.read_csv('50_Startups.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

## Encoding Categorical Data

- The state column needs to be encoded as it is categorical

```python
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
print(X)
```

## Splitting the Dataset into the Training Set and Test Set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

- Coefficients adjust: In multiple linear regression, the model adjusts the coefficients of variables according to their scale, making scaling not strictly necessary.
- Relative scales: The importance of a feature is determined by its coefficient, not its scale.
- Interpretability: Keeping original units makes model outputs easier to interpret in the context of the data.

## Training the Multiple Linear Regression Model on the Training Set

- The `LinearRegression` class can automatically handle simple and multiple linear regression
- Automatically selects the features that are the most statistically significant and automatically handles dummy variable trap for the encoded categorical data

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## Predicting the Test Set Results

```python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
data = np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1)

labels = [f'pred {i+1} & test {i+1}' for i in range(len(data))]

pos = np.arange(len(data))
bar_width = 0.35

fig, ax = plt.subplots()
bar1 = ax.bar(pos - bar_width/2, data[:,0], bar_width, label='Prediction')
bar2 = ax.bar(pos + bar_width/2, data[:,1], bar_width, label='Test')

ax.set_ylabel('Values')
ax.set_title('Predictions vs Test Values')
ax.set_xticks(pos)
ax.set_xticklabels(labels, rotation=90)  # Rotate labels to avoid overlap
ax.legend()

plt.tight_layout()
plt.show()
```
