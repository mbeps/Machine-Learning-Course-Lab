# Simple Linear Regression

## Importing the Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

## Importing the Dataset

```python
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

## Splitting the Dataset into the Training Set and Test Set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

## Training the Simple Linear Regression Model on the Training Set

- Linear regression predicts a dependent variable using one or more independent variables.
- Simple linear regression uses only one independent variable to predict a dependent variable.
- It is used when there is a linear relationship between the independent and dependent variables.
- Mathematical formula: \(y = mx + c\), where:
  - \(y\) is the predicted value,
  - \(m\) is the slope of the line,
  - \(x\) is the independent variable,
  - \(c\) is the y-intercept.
- Ordinary Least Squares (OLS) is a method used in linear regression to estimate the parameters (slope and intercept) by minimizing the sum of the squared differences between the observed and predicted values.

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## Predicting the Test Set Results

- Applying the model onto the testing data 
- Using the matrix of features from the testing data to predict the outcome

```python
y_pred = regressor.predict(X_test)
```

## Visualising the Training Set Results

- The red dots represent the actual points from the training set and the blue line is the prediction

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Test Set Results

- The red dots represent the actual points from the testing set and the blue line is the prediction
- The line should be close the actual points from the test set showing that the model is good at predicting unseen data

```python
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
